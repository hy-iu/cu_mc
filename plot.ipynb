```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
positions = np.random.rand(1000, 3)*5
colors = np.zeros(1000)
x00 = x01 = x10 = x11 = True
pos00 = np.zeros(2)
pos01 = np.zeros(2)
pos10 = np.zeros(2)
pos11 = np.zeros(2)
for i in range(len(positions)):
    if x00 and 2 < positions[i, 0] < 3 and 2 < positions[i, 1] < 3:
        colors[i] = 1
        pos00 = positions[i, :2]
        x00 = False
    elif x01 and not x00 and 2 < positions[i, 0] < 3 and 2 < positions[i, 1] < 3:
        colors[i] = 1
        pos01 = positions[i, :2]
        x01 = False
    elif x10 and not x00 and not x01 and 2 < positions[i, 0] < 3 and 2 < positions[i, 1] < 3:
        colors[i] = 1
        pos10 = positions[i, :2]
        x10 = False
    elif x11 and 1 < positions[i, 0] < 2 and 2 < positions[i, 1] < 3:
        colors[i] = 1
        pos11 = positions[i, :2]
        x11 = False

plt.figure(figsize=(10, 10))
plt.scatter(positions[:, 0], positions[:, 1], s=3, c=colors, cmap='coolwarm')
plt.plot([pos00[0], pos01[0]], [pos00[1], pos01[1]], 'r--', linewidth=1)
plt.plot([pos10[0], pos11[0]], [pos10[1], pos11[1]], 'b--', linewidth=1)
plt.grid(True)
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.xticks(ticks=[0, 1, 2, 3, 4, 5],labels=[])
plt.yticks(ticks=[0, 1, 2, 3, 4, 5],labels=[])
plt.xlabel('x')
plt.ylabel('y')

plt.fill([2, 2, 3, 3], [2, 3, 3, 2], 'r', alpha=0.3)

plt.fill([1, 1, 2, 2], [2, 3, 3, 2], 'b', alpha=0.3)
plt.fill([2, 2, 3, 3], [1, 2, 2, 1], 'b', alpha=0.3)
plt.fill([2, 2, 3, 3], [3, 4, 4, 3], 'b', alpha=0.3)
plt.fill([3, 3, 4, 4], [2, 3, 3, 2], 'b', alpha=0.3)

plt.fill([0, 0, 1, 1], [2, 3, 3, 2], 'b', alpha=0.1)
plt.fill([1, 1, 2, 2], [1, 2, 2, 1], 'b', alpha=0.1)
plt.fill([1, 1, 2, 2], [3, 4, 4, 3], 'b', alpha=0.1)
plt.fill([2, 2, 3, 3], [0, 1, 1, 0], 'b', alpha=0.1)
plt.fill([2, 2, 3, 3], [4, 5, 5, 4], 'b', alpha=0.1)
plt.fill([3, 3, 4, 4], [1, 2, 2, 1], 'b', alpha=0.1)
plt.fill([3, 3, 4, 4], [3, 4, 4, 3], 'b', alpha=0.1)
plt.fill([4, 4, 5, 5], [2, 3, 3, 2], 'b', alpha=0.1)
```
